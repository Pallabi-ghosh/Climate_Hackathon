```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.rc("font", size=14)


import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
survey=pd.read_csv('data.csv',index_col=0) #Read the dataset
```

```python
survey['Gender'].unique() #print the unique values in Gender column
```

```python
sns.countplot(x='Gender',palette='hls',data=survey)
```

```python
#Histogram to show the relationship between 'age' and 'Gender'
sns.barplot(x ='Gender', y ='Age', data = survey, palette ='plasma')
plt.xticks(rotation=45)
```

```python
survey.drop(['Name','Place','Interview Date '],inplace=True, axis=1) #Drop unwanted columns
```

```python
categorical_features=[feature for feature in survey.columns if ((survey[feature].dtypes=='O') & (feature not in ['Gender']))]
categorical_features
```

```python
#check count based on categorical features
plt.figure(figsize=(15,80), facecolor='white')
plotnumber =1
for categorical_feature in categorical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(y=categorical_feature,data=survey)
    plt.xlabel(categorical_feature)
    plt.title(categorical_feature)
    plotnumber+=1
plt.show()
```
